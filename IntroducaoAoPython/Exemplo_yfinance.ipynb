## Construção de Carteira e comparação com o IBOV

```python
ativos = ['PETR4','VALE3','ABEV3','WEGE3','EMBR3','RADL3','CVCB3','CIEL3','OIBR3','KNRI11']
```

```python
inicio='2020-01-01'
fim='2021-08-15'
```

```python
def retorna_ativos(ativos, inicio, fim):
  df = pd.DataFrame()
  for i in ativos:
    ticker = str(i) + '.SA'
    df[i] = yf.download(ticker, start=inicio, end=fim)['Adj Close']
  
  return df
```

```python
precos = retorna_ativos(ativos,inicio,fim)
```

```python
precos.head()
```

## Quantos papéis compramos de cada um desses ativos?

```python
primeira = precos.iloc[0]
```

```python
primeira
```

```python
comprados = round(1000/primeira, 0)
```

```python
comprados
```

## Calcular o patrimonio liquido

```python
PL = precos*comprados
```

```python
PL.head()
```

```python
PL.tail()
```

```python
PL['Minha Carteira'] = PL.iloc[:].sum(axis=1)
```

```python
PL.head()
```

```python
PL.tail()
```

## Baixando os dados do IBOV para fazer comparacao

```python
ibov = yf.download('^BVSP', start = inicio, end = fim)
```

```python
ibov.head()
```

```python
ibov.rename(columns = {'Adj Close': 'IBOV'}, inplace = True)
```

```python
ibov = ibov.drop(ibov.columns[[0,1,2,3,5]], axis = 1)
```

```python
ibov.head()
```

```python
PL.index = pd.to_datetime(PL.index)
```

## Merge dos dataframes ibov(indice BOVESPA) e PL(Patrimonio Liquido)

```python
novo_df = pd.merge(ibov, PL, how='inner', on='Date')
```

```python
novo_df.head()
```

## Normalizado

```python
normalizado = novo_df/novo_df.iloc[0]
```

```python
normalizado.head()
```

```python
normalizado[['IBOV', 'Minha Carteira']].plot(figsize = (10,10));
```
