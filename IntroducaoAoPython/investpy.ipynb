## como pesquisa do site investing.com

```python
search_results = investpy.search_quotes(text='PETR', products=['stocks'], countries=['brazil'], n_results=50)
```

```python
for search_result in search_results:
  print(search_result)
```

```python
for search_result in search_results[1:2]:
  print(search_result)
```

```python
petr = search_result.retrieve_historical_data(from_date='01/01/2014', to_date='01/06/2021')
```

```python
# OHLC - Open, High, Low, Close
petr.head()
```

```python
petr['Close'].plot()
```
