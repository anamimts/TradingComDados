## Conceito de Normalização

```python
lista = ['WEGE3.SA', 'MGLU3.SA', 'EMBR3.SA', '^GSPC', '^BVSP', 'HCTR11.SA', 'SMAL11.SA']
```

```python
import pandas as pd
```

```python
data = pd.DataFrame()
```

```python
for i in lista:
  data[i] = pdr.data.get_data_yahoo(i, start='2019-01-01').Close
```

```python
data.head()
```

```python
data.shape
```

```python
data_sem_na = data.dropna()
```

```python
data_sem_na.head()
```

```python
data_sem_na.shape
```

```python
import matplotlib.pyplot as plt
```

```python
data_sem_na.plot(figsize = [8,8]);
plt.title('Ativos sem normalizacao', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Preco')
plt.show()
```

```python
data_sem_na.head()
```

```python
normalizado = data_sem_na/data_sem_na.iloc[0]
```

```python
normalizado.head()
```

```python
normalizado.plot(figsize = [8,8]);
plt.title('Ativos normalizados a 1', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Preco')
plt.show()
```
