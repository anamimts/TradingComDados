## 15. kwargs

```python
def compra_acoes(**kwargs):
  if 'acao' in kwargs:
    print('a ação comprada foi ', kwargs['acao'])
  else:
    print('O cliente não comprou ações')
```

```python
compra_acoes(acao = 'PETR4')
```

```python
def montagem_posicao(*args, **kwargs):
  if 'acao' and 'lote' in kwargs:
    print('O cliente montou uma posição de ', ' e '.join(kwargs['acao']))
    print('Comprou ', ' e '.join(args), 'papéis, respectivamente')
    print('Todas elas foram compradas no lote ', kwargs['lote'])
```

```python
montagem_posicao('100', '200', acao = ('PETR4', 'WEGE3'), lote = 'inteiro')
```
